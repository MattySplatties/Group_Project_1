```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import numpy as np

warnings.filterwarnings('ignore')
```

```python
FB_file = "Each_SM_data_set/FB_data.csv"
FB_df = pd.read_csv(FB_file)
Twitter_file = "Each_SM_data_set/Twitter_data.csv"
Twitter_df = pd.read_csv(Twitter_file)
Insta_file = "Each_SM_data_set/Instagram_data.csv"
Insta_df = pd.read_csv(Insta_file)
Youtube_file = "Each_SM_data_set/Youtube_data.csv"
Youtube_df = pd.read_csv(Youtube_file)
Reddit_file = "Each_SM_data_set/Reddit_data.csv"
Reddit_df = pd.read_csv(Reddit_file)
Disc_file = "Each_SM_data_set/Discord_data.csv"
Disc_df = pd.read_csv(Disc_file)
Pin_file = "Each_SM_data_set/Pintrest_data.csv"
Pin_df = pd.read_csv(Pin_file)
Tiktok_file = "Each_SM_data_set/Tiktok_data.csv"
Tiktok_df = pd.read_csv(Tiktok_file)
```

```python
FB_time = FB_df.groupby(["Average time on social media (per Day)"])["How often do you feel depressed or down?"].mean()
Twitter_time = Twitter_df.groupby(["Average time on social media (per Day)"])["How often do you feel depressed or down?"].mean()
Insta_time = Insta_df.groupby(["Average time on social media (per Day)"])["How often do you feel depressed or down?"].mean()
Youtube_time = Youtube_df.groupby(["Average time on social media (per Day)"])["How often do you feel depressed or down?"].mean()
Reddit_time = Reddit_df.groupby(["Average time on social media (per Day)"])["How often do you feel depressed or down?"].mean()
Disc_time = Disc_df.groupby(["Average time on social media (per Day)"])["How often do you feel depressed or down?"].mean()
Pin_time = Pin_df.groupby(["Average time on social media (per Day)"])["How often do you feel depressed or down?"].mean()
Tiktok_time = Tiktok_df.groupby(["Average time on social media (per Day)"])["How often do you feel depressed or down?"].mean()
```

```python
FB_time = FB_time.reindex(["Less than an Hour","Between 1 and 2 hours","Between 2 and 3 hours","Between 3 and 4 hours","Between 4 and 5 hours","More than 5 hours"])
Twitter_time = Twitter_time.reindex(["Less than an Hour","Between 1 and 2 hours","Between 2 and 3 hours","Between 3 and 4 hours","Between 4 and 5 hours","More than 5 hours"])
Insta_time = Insta_time.reindex(["Less than an Hour","Between 1 and 2 hours","Between 2 and 3 hours","Between 3 and 4 hours","Between 4 and 5 hours","More than 5 hours"])
Youtube_time = Youtube_time.reindex(["Less than an Hour","Between 1 and 2 hours","Between 2 and 3 hours","Between 3 and 4 hours","Between 4 and 5 hours","More than 5 hours"])
Reddit_time = Reddit_time.reindex(["Less than an Hour","Between 1 and 2 hours","Between 2 and 3 hours","Between 3 and 4 hours","Between 4 and 5 hours","More than 5 hours"])
Disc_time = Disc_time.reindex(["Less than an Hour","Between 1 and 2 hours","Between 2 and 3 hours","Between 3 and 4 hours","Between 4 and 5 hours","More than 5 hours"])
Pin_time = Pin_time.reindex(["Less than an Hour","Between 1 and 2 hours","Between 2 and 3 hours","Between 3 and 4 hours","Between 4 and 5 hours","More than 5 hours"])
Tiktok_time = Tiktok_time.reindex(["Less than an Hour","Between 1 and 2 hours","Between 2 and 3 hours","Between 3 and 4 hours","Between 4 and 5 hours","More than 5 hours"])
```

```python
FB = FB_time.plot(kind='bar', title='Time spent of Facebook vs. Average of how depressed individuals are',
               ylabel='How often do you feel depressed or down?', xlabel='Average time on social media (per Day)', figsize=(10, 6), color = ((66/255),(103/255),(178/255)))
plt.savefig("Data_Images/FB_TimevsDep")
```
